### Background  <a id='section-2'>
This can have literature review, articles, etc.

According to the U.S. Department of Agriculture (USDA), food insecurity is defined as the lack of consistent access to enough food for an active, healthy life1. Food insecurity affects all communities in the U.S.; however, seniors, children, Latinos, African Americans, and rural communities are disproportionately impacted. In 2019, an estimated 10.5% of all U.S. households (equivalent to 35 million Americans) were food insecure for at least some part of the year2. Because these numbers reflect data collected in December 2019, prior to the impact of COVID-19 in 2020, it is estimated that the 2020 statistics for food insecurity will be even higher [2]. Feeding America predicted in October 2020 that the projected annual food insecurity rate for the year would be approximately 15.6% in the overall U.S. population, with a projected 50.4 million Americans experiencing food insecurity at some point in the year [2].

Historically, food assistance programs such as the Supplemental Nutrition Assistance Program (SNAP) and Special Supplemental Program for Women, Infants, and Children (WIC) have been utilized in order to enable low-income families with the ability to purchase nutritious food [3]. In FY2020, The Emergency Food Assistance Program (TEFAP) received a significant amount of additional funding for food purchases and administrative expenses through the Food Purchase Distribution Program (FPDP) and legislation in response to the novel coronavirus public health emergency [3]. With consideration to the COVID-19 pandemic and rising food insecurity, Congress’ Families First Coronavirus Act (FFCA) enabled states to issue pandemic electronic meal-replacement benefits (P-EBT) for households with children eligible to receive free or reduced-price school meals, which amounts to a dollar value of roughly $114 per child a month [4]. During school closures, local educational authorities are also enabled to use feeding practices from the USDA’s summer meals programs to ensure children receive adequate nutrition [4]. Additionally, individual states are allowed to increase SNAP allotments up to the maximum benefit amount as an emergency measure [4]. 

With consideration to the rising concern around COVID-19, there is a parallel rising food insecurity [4]. Prior to the pandemic, the number of Americans experiencing food insecurity had been steadily falling for the past few years. However, the increase in job loss due to the pandemic has caused many financial struggles for Americans, including putting food on the table. Communities that already had high rates of food insecurity prior to the pandemic have seen even higher rates, while those that had lower rates before have seen larger relative increases in food insecurity [5]. Given reduced public transportation, communities existing in food deserts face increased geographic barriers to accessing food, and pandemic driven shifts in food shopping habits such as “food hoarding” leave low-income families who can not afford to buy food in bulk at a disadvantage [4]. Groceries in food deserts also cost more compared to those in suburban markets, which impacts low-income communities further. Due to the economic instability caused by the pandemic, states are seeing the largest absolute increase in food-insecure children, Latino, Native American, and Black communities [5]. Many children rely on school lunch programs and with schools being closed down by the pandemic, these children no longer have a reliable source for food. In addition, food insecurity disproportionately affects the Latino, Native American, and Black communities because of higher-than-average unemployment and poverty rates caused by histories of systemic racism and socioeconomic disparity [6]. Overall, the COVID-19 pandemic has contributed significantly to the rise in national and state food insecurity rates, while disproportionately affecting lower income communities.

As many government assistance programs and federal policy initiatives work to address the rapidly rising food insecurity due to COVID-19, there is a patchwork response to food insecurity on state and local levels [4]. At a state level, 2.5 million households with children already receive the maximum benefit and states are not allowed to increase benefit levels beyond that maximum [4]. Geographic variance in the number of free food resources such as food banks and soup kitchens provided by each state can also present a challenge for states with increasing population-level food insecurity. Additionally, despite local efforts to reduce child food insecurity through school feeding programs and summer meal programs, federal guidelines for executing feeding programs safely amidst the COVID-19 outbreak are lacking, and many school districts do not have experience with summer feeding programs [3]. These differences between food program operations at state and local levels in combination with lacking policy efforts and pandemic-driven barriers to food accessibility must be addressed to mitigate rapidly increasing food insecurity levels exacerbated by the COVID-19 pandemic.

### Methodology <a id='section-4'>
Explain the relative search volume, the process of finding the sample of top queries, choosing terms, etc.

We obtain key pieces of Google search data by using two Google APIs. Given an initial search term for a specified location and time period, Google Trends provides the top search topics and top search queries with a Relative Search Index (RSI). Google Health Trends generates a normalized Relative Search Volume (RSV) for a list of top queries in a specific region and time period. These two metrics differ in that the RSI provides a search interest relative to all searches during the specified period of time, whereas the normalized RSV refers to the proportion of searches for a specific query as compared to the sum total of searches for a set of queries, allowing for more in-depth analysis of the relationships between queries.

Initially, we embarked on our exploratory analysis by observing interesting trends and features of nutrition-related search terms in the public Google Trends API. We looked at different terms related to pregnancy and nutrition, prenatal supplements, and food insecurity. We noted our qualitative and quantitative observations and subsequently used the Google Health Trends API to view more descriptive analytical figures including RSV and Top Queries. Specific qualities we assessed were the trends in RSV for terms over a one year period (2019) as well as over a five year period (2015-2019). We also looked at the number and quality of related top queries as it relates to nutrition and user-centric questions. We narrowed in on terms that showed relatively large increases in RSV for the respective time period and had a high volume of related top queries. From our exploratory research, we noted that food insecurity was a particular field that showed the most promise in its initial terms and we felt could be best addressed using Google API data. We wanted to view how initial search terms differed in their temporal and geographic patterns across the United States. Since it would be too time-intensive and repetitive to view these patterns for all 50 states, we chose one representative state from each of the 7 USDA regions to perform our analysis on. The seven states selected (aside from California) were all intentionally characterized with a low mean income based off of the IPUMS-CPS income data and is depicted in [this chart](https://dqydj.com/average-income-by-state-median-top-percentiles/). We also wanted to make sure that our final list of terms were commonly searched within these states and covered multiple facets of food insecurity. 

As such, we identified 12 search terms (EBT, WIC, Snap/Snap food, Food coupons, Food assistance, Food banks, Food stamps, Free food, Food shortage, Emergency food, Free meal) related to food insecurity that we felt were the most insightful moving forward using our specific metrics. 

Our original concern when using the Google Trends API was the factors that go into deciding whether or not a term is placed within the “health” category. While food insecurity is definitely a public health issue, it was unclear whether or not the Google API would classify this strictly under the “health” category and whether or not this would tamper with its results. We sought information via Google’s API guide and on the internet but found no answers to our question. As such, we looked at the RSV and related queries quality of our terms using both the “general” and “health” categories on Google Trends. However, in deciding to narrow down our original list of 12 terms to 5 terms, we ultimately decided on using the “general” category as our standard since it provided a more comprehensive set of related search queries (ie: p-ebt, applying, benefits, etc.) compared to just the “Health Category” (ie: wic health) while also covering most if not all of the results from the “health” category. The only exception is “SNAP”, as the related queries for all categories included other extraneous topics, such as snapchat, snap camera, snap peas, etc.), which we used “SNAP food” for to combat these irrelevant topics. Following this subtle note, we used many of the same metrics as described in the previous step: we looked at RSV temporal patterns over the past year (2020), the RSV geographical patterns as distributed across the 7 states, as well as the quality and quantity of related queries. We immediately threw out terms that did not show enough data for related queries and used the median RSV from the past year to decide which terms were the most commonly searched. We also looked at how many of the 7 states showed enough data for these terms, and used these holistic median values for each term across all states to capture a better picture of how these terms fare across the U.S. Furthermore, we assessed the quality and quantity of the related queries, noting whether or not any related queries provided additional information or insight to the original term.
